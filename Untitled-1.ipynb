```python
from kagglehub import KaggleDatasetAdapter
import matplotlib.pyplot as plt
import pandas as pd
import math
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression, RidgeCV
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


df = pd.read_excel("data.xlsx")
```

Целевая переменная - цена за электричество. Факторы: площадь учатска, потребление воды, индекс качества воздуха, количество проживающих, коэффициент использования

```python
columns = ['electricity cost', 'site area', 'water consumption', 'air qality index', 'resident count']


def build_histo(bins):
    for i in columns:
        df[i].hist(bins=int(bins), edgecolor='black')
        plt.title(i)
        plt.xlabel("Значения")
        plt.ylabel("Частота")
        plt.show()


print("Мода: \n", df.mode(), "\n")
print("Минимальные значения \n", df.min(), "\n")
print("Максимальные значения \n", df.max(), "\n")
print('Эксцесс\n', df.kurt(), "\n")
print('Медиана\n', df.median(), "\n")
print('Среднее\n', df.mean(), "\n")
print('Ст отклонение\n', df.std(), "\n")
print('Дисперсия\n', df.var(), "\n")
print('Ассиметрия\n', df.skew(), "\n")

k = 1 + 1.322*math.log10(len(df['electricity cost']))
build_histo(k)
```

Проведён дискриптивный анализ переменных, построены гистограммы.
По данным и гистограммам можно заметить, что распределение стоимости электричества похоже на нормальное, потребление воды и количество жильцов на экспоненциальное, площадь и качетсво воздуха - на равномерное.
Эксцесс у количества жильцов -0.94, ассиметрия 1.31, что говорит об экстремальных значениях и большом хвосте. Эксцесс потребления воды -0.17, ассиметрия - 0.77, распределение близкое к нормальному.

```python
def normalize_dataset():
    normalized_df = (df - df.mean()) / df.std()
    return normalized_df


z_normalized_df = normalize_dataset()
```

Предобработка данных. Пропущенных значений нет. Нормализуем данные через z-нормализацию. Далее нормализация не даст сильных изменений.

```python
cor_matrix = z_normalized_df.corr()
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()

X_const = sm.add_constant(z_normalized_df)
vif_data = pd.DataFrame()
vif_data['feature'] = X_const.columns
vif_data['VIF'] = [
    variance_inflation_factor(X_const.values, i)
    for i in range(X_const.shape[1])
]
print(vif_data)
```

Видно, что качество воздуха и коэффициент утилизации почти ни с чем не коррелируют. Сильную корреляцию можно заметить в Площадь-Стоимость, Площадь-Вода, Вода-Стоимость.
VIF коэффициента показали, что у Площади может присутствовать мультикорреляция. 

```python
def regr(dataframe, target='electricity cost'):
    X = dataframe.drop(columns=[target])
    Y = dataframe[target]
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    modRidge = RidgeCV(alphas=[0.1, 1, 5, 10, 50, 100])
    modLinear = LinearRegression()
    models = {'Линейная': modLinear, 'Гребневая': modRidge}
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)

    for name, model in models.items():
        model.fit(X_train, y_train)
        predicted = model.predict(X_test)

        rmse = root_mean_squared_error(y_test, predicted)
        r2 = r2_score(y_test, predicted)
        mape = mean_absolute_percentage_error(y_test, predicted)
        print(f"{name} регрессия")
        print('RMSE: ', rmse)
        print('R2: ', r2)
        print('MAPE: ', mape)

        cv_scores = cross_val_score(model, X, Y, scoring='r2', cv=kfold)
        print(f"Кросс-валидация R2 (среднее): {cv_scores.mean():.4f}")
        print(f"R2 по фолдам: {cv_scores}\n")

        plt.figure(figsize=(8,6))
        plt.scatter(y_test, predicted, alpha=0.5)
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
        plt.xlabel('Фактические значения')
        plt.ylabel('Предсказанные значения')
        plt.title(f'{name} регрессия (множественная)')
        plt.grid(True)
        plt.show()
regr(z_normalized_df)
```

Множественная линейная регрессия. Разделим выборку 80/20. Построим линейную и гребневую модели(гребневая модель со встроенной кросс-валидацией RidgeCV сама вычисляет подходящий параметр alpha), подсчитаем ошибки и коэффициент детерминации. Подсчитаем R2, использую кросс-валидацию. Видно, что кросс-валидация показала почти тот же коэффициент в обоих случаях. Судя по метрикам, модели мало отличаются.

```python
def MGP():
    scaler = StandardScaler()
    scaled = scaler.fit_transform(df)
    X = df.drop(columns=['electricity cost'])
    pca = PCA()
    pca.fit(scaled)

    explained_variance_ratio = pca.explained_variance_ratio_
    cum_explained_variance = explained_variance_ratio.cumsum()

    plt.figure(figsize=(8,6))
    plt.plot(range(1, len(cum_explained_variance)+1), cum_explained_variance, marker='o')
    plt.xlabel('Количество главных компонент')
    plt.ylabel('Накопленная доля объяснённой дисперсии')
    plt.title('Метод локтя для PCA')
    plt.grid(True)
    plt.axhline(0.9, color='red', linestyle='--', label='90% дисперсии')
    plt.legend()
    plt.show()
    pca = PCA(n_components=0.9)
    principal_components = pca.fit_transform(scaled)

    print("Сколько компонент оставлено:", pca.n_components_)
    col_names = [f'Col_{i+1}' for i in range(principal_components.shape[1])]
    pca_df = pd.DataFrame(principal_components, columns=col_names)
    pca_df['electricity cost'] = df['electricity cost']
    col_names.append('electricity cost')

    regr(pca_df)
MGP()
```

Применим метод главных компонент, позволяющий убрать мультиколлениарность и понизить размерность модели. Ситуация с разными моделями(линейная и гребневая) та же: метрики почти не отличаются. По графику локтя видно, что лучше взять 4 компоненты из 5(чтобы суммарный вклад компонент был не менее 90%). Размерность уменьшена. Метрики улучшились. Так, коэффициент детерминации R2 увеличился на 0.08 по сравнению с п.4 лабораторной работы.
